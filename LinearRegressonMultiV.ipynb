```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples and features
num_samples = 100
num_features = 3

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, num_features)

# Define true weights (slopes) and intercept
true_weights = np.array([[3], [2], [1]])  # shape: (3, 1)
true_intercept = 4

# Generate noise
noise = np.random.randn(num_samples, 1)

# Compute y values
y_train = true_intercept + x_train @ true_weights + noise  # @ is matrix multiplication

# Plotting using only first feature for visualization
plt.figure(figsize=(8, 5))
plt.scatter(x_train[:, 0], y_train, color="green", label="Training data (vs 1st feature)", alpha=0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:",y_train.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x_train and y_train first
# For example, you could create some sample data:
np.random.seed(42)
x_train = np.random.rand(100, 3)  # 100 samples, 3 features
y_train = 2*x_train[:, 0] + 3*x_train[:, 1] + 4*x_train[:, 2] + 1 + np.random.randn(100)*0.1
b = 0
alpha = 0.001
cost = 0

allCost = []
m = x_train.shape[0]
w = np.array([[1.1], [2.1], [3.1]])
for i in range(500):
    y_pred = x_train @ w + b
    cost = np.sum((y_train - y_pred)**2)/m
    allCost.append(cost)
    w = w - alpha * (1/m) * x_train.T @ (y_pred - y_train)
    b = b - alpha * (1/m) * np.sum(y_pred - y_train)
plt.figure(figsize=(8, 5))
plt.scatter(x_train[:, 0], y_train, color="green", label="Training data (vs 1st feature)", alpha=0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.plot(allCost)
print(allCost[-1])
plt.show()
```
